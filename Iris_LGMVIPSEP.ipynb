## Processing and Splitting the Data


### Model Building 


### x is having all the dependent variable  and y is having all the independent variable (in this case class is independent vaiable)

```python
x = data.drop(['Class'],axis =1)
y = data['Class']
```

```python
x.shape 
```

```python
y.shape
```

### TrainTest Split

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, random_state=0, test_size =0.10)
```

### Model creation

```python
model = []
model.append(('SVC',SVC(gamma='auto')))
```

```python
model = SVC(gamma='auto')
model.fit(x_train,y_train)
prediction = model.predict(x_test)
```

### Now checking the accuracy of model

```python
print(accuracy_score(y_test,prediction))
```

```python
print(classification_report(y_test,prediction))
```

## Conclusion

### For this model the Accuracy on the test set is 1.00 which means that model made the right prediction for 100% irises in the dataset.This is a highlevel of accuracy and it means that model may be trustworthy for use.
